```python
numerical_cols = ['finalWorth', 'age', 'cpi_country', 'cpi_change_country', 'population_country']
corr_matrix = df[numerical_cols].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# wealth gap by country
gender_wealth_gap = df.groupby(['country', 'gender'])['finalWorth'].mean().unstack().drop(columns=['Unknown']).dropna()
gender_wealth_gap['wealth_gap'] = gender_wealth_gap['M'] - gender_wealth_gap['F']
gender_wealth_gap['wealth_gap'].dropna().sort_values().plot(kind='bar', figsize=(15, 7), color='skyblue')
plt.title('Gender Wealth Gap by Country')
plt.xlabel('Country')
plt.ylabel('Average Net Worth Difference (Male - Female)')
plt.xticks(rotation=45)
plt.show()
```

```python
tertiary_corr = stats.pearsonr(df['gross_tertiary_education_enrollment'], df['finalWorth'])
primary_corr = stats.pearsonr(df['gross_primary_education_enrollment_country'], df['finalWorth'])

print(f"Overall Pearson Correlation between tertiary education enrollment and net worth: {tertiary_corr[0]} (p-value: {tertiary_corr[1]})")
print(f"Overall Pearson Correlation between primary education enrollment and net worth: {primary_corr[0]} (p-value: {primary_corr[1]})")

for gender in df['gender'].unique():
    subset = df[df['gender'] == gender]
    tertiary_corr_gender = stats.pearsonr(subset['gross_tertiary_education_enrollment'], subset['finalWorth'])
    primary_corr_gender = stats.pearsonr(subset['gross_primary_education_enrollment_country'], subset['finalWorth'])
    
    print(f"{gender} Pearson Correlation between tertiary education enrollment and net worth: {tertiary_corr_gender[0]} (p-value: {tertiary_corr_gender[1]})")
    print(f"{gender} Pearson Correlation between primary education enrollment and net worth: {primary_corr_gender[0]} (p-value: {primary_corr_gender[1]})")
```

### Statistical Analysis

We are performing a regression analysis to see if gender is a significant predictor of net worth when other factors are accounted for, and to visualize the relationship between net worth and factors like age, industry, and country while controlling for the gender variable.

```python
# creating one hot encoding for gender
df['gender_ohe'] = df['gender'].map({'female':1, 'male':0})
df = pd.get_dummies(df, columns=['industries', 'country'], drop_first=True)
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(df.mean(), inplace=True)  # Example: Filling with the mean
X = df.drop(['finalWorth', 'personName', 'city', 'source', 'countryOfCitizenship', 'status', 'birthDate', 
             'lastName', 'firstName', 'date', 'state', 'residenceStateRegion', 'birthYear', 'birthMonth', 'birthDay'], axis=1)
X = X.select_dtypes(include='number')
y = pd.DataFrame(df['finalWorth'])
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary)
```
